# 01 — Data Loading

This notebook loads pulsar photon arrival-time data and performs an initial inspection of the dataset.

## Objectives

- load the dataset;
- inspect the available columns;
- check data types and missing values;
- prepare the data for further exploratory analysis.

```python
from pathlib import Path

import numpy as np
import pandas as pd
from astropy.io import fits
```

```python
data_path = Path("../data/raw/vela_photons.fits")

if not data_path.exists():
    raise FileNotFoundError(
        f"Data file not found: {data_path}\n"
        "Please place vela_photons.fits into data/raw/ before running this notebook."
    )

data_path
```

```python
with fits.open(data_path) as hdul:
    hdul.info()
    events = hdul[1].data
```

```python
df = pd.DataFrame({
    "time": np.asarray(events["TIME"], dtype=np.float64),
    "energy": np.asarray(events["ENERGY"], dtype=np.float64),
})

df.head()
```
