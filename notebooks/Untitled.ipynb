```python
This notebook loads the ATNF Pulsar Catalogue and performs initial exploration
to understand pulsar properties relevant for timing stability and navigation.
"""

import psrqpy
import pandas as pd
import matplotlib.pyplot as plt

print("Environment is ready")

query = psrqpy.QueryATNF()
df = query.table.to_pandas()
df.head()

"""
## Initial Observations

- The dataset contains both observational and derived pulsar parameters.
- Columns include: period (P0), period derivative (P1), distance (DIST), RA, DEC, etc.
- Some columns have missing values, which can be handled in preprocessing.
- This dataset provides a suitable basis for exploratory analysis of pulsar timing characteristics.
"""

plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(10,5))
plt.hist(df['P0'], bins=50, color='purple', edgecolor='black')
plt.xlabel('Period (s)')
plt.ylabel('Number of Pulsars')
plt.title('Distribution of Pulsar Periods')
plt.show()

plt.figure(figsize=(10,5))
plt.scatter(df['DIST'], df['P0'], alpha=0.7, color='teal')
plt.xlabel('Distance (kpc)')
plt.ylabel('Period (s)')
plt.title('Pulsar Period vs Distance')
plt.show()

"""
## Conclusions

- Most pulsars have periods between ~0.1 s and ~1 s, but there are also millisecond pulsars with periods <0.01 s.
- Scatter plot shows that pulsars used for navigation purposes can be at varying distances.
- This notebook demonstrates the ability to load real astronomical data,
  perform exploratory data analysis, and produce publication-ready visualizations.
"""
```
